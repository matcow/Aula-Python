## 02 – Altere os operadores relacionais para modificar o resultado de cada uma das comparações.

<br>('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
<br>['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
<br>['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
<br>['a',1,True,'b',0,False] > ['a',True,1]
<br>['a',1,True,'b',0,False] == 1
<br>['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

```python
print("exercicio 01:")
print(('a',1,True,'b',0,False) != ('a',True,1,'b',False,0))
print("substituindo por == resultado em:")
print(('a',1,True,'b',0,False) == ('a',True,1,'b',False,0))

print("exercicio 02:")
print(['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)) 
print("substituindo por == resultado em:")
print(['a',1,True,'b',0,False] == ('a',True,1,'b',False,0))

print("exercicio 03:")
print(['a',1,True,'b',0,False] != ['a',True,1,'b',False,0])
print("substituindo por == resultado em:")
print(['a',1,True,'b',0,False] == ['a',True,1,'b',False,0])

print("exercicio 04:")
print(['a',1,True,'b',0,False] > ['a',True,1])
print("substituindo por < resultado em:")
print(['a',1,True,'b',0,False] < ['a',True,1])

print("exercicio 05:")
print(['a',1,True,'b',0,False] == 1)
print("substituindo por != resultado em:")
print(['a',1,True,'b',0,False] != 1)

print("exercicio 06:")
print(['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False])
print("substituindo por > resultado em:")
print(['a',True,1] > ['a',True,1] < ['a',1,True,'b',0,False])

```
