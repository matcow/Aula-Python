```python
ex_01 = list(range(5000)) 
```

## 03 – Escreva o programa em Python:

```python
pares = 0 
impares = 0

for numero in ex_01:
    if numero %2 == 0:
        pares +=1
    else:
        impares +=1

print('pares:', pares) 
print('impares:', impares)
```
