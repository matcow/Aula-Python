## 04 – Defina uma função em Python que receba uma pd.Series contendo uma variável quantitativa e retorne uma DataFrame contendo a tabela de frequência.

Obtenha a tabela de frequência das variáveis criadas nos exercícios 1 e 2.

```python
def tab_freq(serie: pd.Series) -> pd.DataFrame:
    freq = serie.value_counts().sort_index()
    df_freq = pd.DataFrame({
        "Valor": freq.index,
        "Frequência": freq.values
    })
    return df_freq

freq_expressao = tab_freq(df_02["expressao"])
print("Tabela de frequência - Expressão gênica")
print(freq_expressao.head())

freq_diferencial = tab_freq(df_02["diferencial"])
print("\nTabela de frequência - Genes diferencialmente expressos")
print(freq_diferencial.head())
```

## 05 – Gere os gráficos de barras e linhas para as frequências em cada umas das variáveis.

```python
bins = [-np.inf, -3, -2, 0, 2, 3, np.inf]
labels = [
    "Muito downregulated",
    "Moderado downregulated",
    "Leve downregulated",
    "Leve upregulated",
    "Moderado upregulated",
    "Muito upregulated"
]

df_02["grupo_expressao"] = pd.cut(df_02["expressao"], bins=bins, labels=labels)

freq_expressao = df_02["grupo_expressao"].value_counts().sort_index()
```

### Expressão

```python
freq_expressao.plot(kind="bar", title="Frequência - Expressão Gênica (categorias)", color="skyblue")
plt.ylabel("Número de genes")
plt.show()

# Gráfico de linhas
freq_expressao.plot(kind="line", marker="o", title="Frequência - Expressão Gênica (categorias)", color="orange")
plt.ylabel("Número de genes")
plt.show()
```
