## 06 – Escreva o programa em Python:

```python
freq = df_ex01['n_caracteres'].value_counts().sort_index()

freq_df = freq.reset_index()
freq_df.columns = ['n_caracteres', 'frequencia']

freq_df
```

```python
freq_df.plot(x='n_caracteres', y='frequencia', kind='bar')
```

## 10 – Obtenha dos seus dados o gráfico abaixo, adicionando 3 replicas.

```python
replicas = []
for i in range(3):
    ruido = np.random.normal(0, 0.5, df_ex07.iloc[:,1:].shape)
    replicas.append(df_ex07.iloc[:,1:].values + ruido)

media = np.mean(replicas, axis=0)
desvio = np.std(replicas, axis=0)

media = np.mean(replicas, axis=0)
desvio = np.std(replicas, axis=0)

fig, ax = plt.subplots(figsize=(10,6))
x = np.arange(len(df_ex07.index))  # posição dos genes
largura = 0.12

for i in range(media.shape[1]):  # para cada amostra
    ax.bar(x + i*largura, media[:,i], yerr=desvio[:,i], width=largura, label=f"Amostra {i+1}")

ax.set_xticks(x + largura*3)
ax.set_ylabel("Valores")
ax.set_title("Gráfico com 3 réplicas")
ax.legend()
plt.show()
```
