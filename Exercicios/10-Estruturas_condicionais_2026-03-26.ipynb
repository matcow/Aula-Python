```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

print (p, q, r, s, t)
falsos=[p,q,r,s,t].count(False)

if p and q and r and s and t:
    print('verdadeiro') 
elif falsos >= 2:
    print('pelo menos dois falsos')
```
