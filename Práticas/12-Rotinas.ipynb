```python
def contar_inteiros(obj: list | tuple) -> int:
    contador = 0
    for x in range(len(obj)):
        if type(obj[x]) == int:
            contador += 1
            print(f'quantidade de inteiros: {contador}')
    return contador
 
```

```python
resultado_contador = contar_inteiros(obj=[5])
print(resultado_contador )
```
