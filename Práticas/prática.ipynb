```python
import pygame
from pygame.locals import *
from OpenGL.GL import *
from OpenGL.GLU import *
import random

pygame.init()
display = (800, 600)
pygame.display.set_mode(display, DOUBLEBUF | OPENGL)

glEnable(GL_DEPTH_TEST)

gluPerspective(45, (display[0]/display[1]), 0.1, 50.0)
glTranslatef(0, 0, -15)  # afasta a câmera pra caber o boneco maior

def cube(size=1):
    s = size / 2
    vertices = [
        [-s,-s,-s],[ s,-s,-s],[ s, s,-s],[-s, s,-s],
        [-s,-s, s],[ s,-s, s],[ s, s, s],[-s, s, s]
    ]

    faces = [
        (0,1,2,3),
        (4,5,6,7),
        (0,1,5,4),
        (2,3,7,6),
        (0,3,7,4),
        (1,2,6,5)
    ]

    glBegin(GL_QUADS)
    for face in faces:
        glColor3f(random.random(), random.random(), random.random())
        for v in face:
            glVertex3fv(vertices[v])
    glEnd()

def steve():
    # 👇 AQUI está o segredo: escala geral
    glPushMatrix()
    glScalef(1.8, 1.8, 1.8)

    # cabeça
    glPushMatrix()
    glTranslatef(0, 2.5, 0)
    glScalef(1.2, 1.2, 1.2)
    cube()
    glPopMatrix()

    # corpo
    glPushMatrix()
    glTranslatef(0, 0.8, 0)
    glScalef(1.5, 2.5, 0.8)
    cube()
    glPopMatrix()

    # braço esquerdo
    glPushMatrix()
    glTranslatef(-1.8, 0.8, 0)
    glScalef(0.5, 2.5, 0.5)
    cube()
    glPopMatrix()

    # braço direito
    glPushMatrix()
    glTranslatef(1.8, 0.8, 0)
    glScalef(0.5, 2.5, 0.5)
    cube()
    glPopMatrix()

    # perna esquerda
    glPushMatrix()
    glTranslatef(-0.5, -2, 0)
    glScalef(0.7, 2, 0.7)
    cube()
    glPopMatrix()

    # perna direita
    glPushMatrix()
    glTranslatef(0.5, -2, 0)
    glScalef(0.7, 2, 0.7)
    cube()
    glPopMatrix()

    glPopMatrix()

angle = 0
clock = pygame.time.Clock()

while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()

    glClear(GL_COLOR_BUFFER_BIT | GL_DEPTH_BUFFER_BIT)

    glPushMatrix()
    glRotatef(angle, 0, 1, 0)

    steve()

    glPopMatrix()

    pygame.display.flip()
    clock.tick(60)

    angle += 1
```
